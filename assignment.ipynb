```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([17, 29, 53, 82])
y3 = np.array([10, 15, 20, 25])
plt.title("Class Size on the disk")
plt.xlabel("Year", fontsize=10,
           family="Arial",
           fontweight="bold",
           color="#2dbefc")
plt.ylabel("Number of participation")
line_style = dict(marker=".",
                markersize=10,
                markerfacecolor="#1cd3fc",
                linestyle="solid",
                linewidth=2)
plt.plot(x,y1,color="#1c5bfc",**line_style)
plt.plot(x,y2,color="#1cfc45",**line_style)
plt.plot(x,y3,color="#fc1c1c",**line_style)
plt.xticks(x)
plt.show()
```

```python
#grid() = Helps make plots easier to read by adding refernece lines.
x=[1,2,3,4,5]
y =[5,11,17,26,29]
plt.grid()
plt.plot(x,y)
plt.show()
```

```python
#bar chart = compare categories of data by representing each category with a bar
categories = ["Grains", "Fruit", "vegatables","protein","Dairy", "Sweets"]
values = np.array([4,3,2,5,3,1])

plt.bar(categories, values, color="skyblue")
plt.xlabel("Food")
plt.ylabel("Quantity")
plt.show()
```

```python
#pie chart
categories =["Freshmen", "shopmores","Juniors", "Seniors"]
values = np.array([300,250,275, 225])
colors=["red","yellow","blue","green"]
plt.pie(values, labels=categories,
        autopct="%1.1f%%",
        colors=colors,
        explode=[0, 0, 0, 0.1],
        shadow=True,
        startangle=90)
plt.title("This is pie chart of some data")
plt.show()
```

```python
#scatter graph = shows the realation between two variables
#                 Helps to identify a correlation (+, -, None)
#                 Example: study hours vs. Test scores

x1 = [0,1,1,2,3,4,5,6,7,7,8]
y1 = [55,60,65,62,68,70,75,78,82,85,90]
x2 = [0,1,1,2,3,4,5,6,7,8,8]
y2 = [45,50,55,72,68,70,72,80,82,85,90]

plt.scatter(x1,y1,color="skyblue",
            alpha = 0.5)
plt.scatter(x2,y2,color="red",
            alpha = 0.5)

plt.title("Scatter Plot")
plt.xlabel("Study Hours")
plt.ylabel("Grade")

plt.legend()
plt.show()
```

```python
#Histograms = A visual repersentation of the distribution of quantites data
#             They group values into bins(intervals)
#             and counts how many fail in each range

scores = np.random.normal(loc=80, scale=10, size=100)
scores = np.clip(scores, 0, 100)
plt.hist(scores, bins=10,
         color="lightgreen",
         edgecolor="black")
plt.title("Exam Scores")
plt.xlabel("score")
plt.ylabel("# of students")
plt.show()
```
